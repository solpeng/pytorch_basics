```python
import torch
import matplotlib.pyplot as plt

x_lst = torch.Tensor([1, 3, 5, 7, 9])
y_lst = torch.Tensor([100, 75, 60, 25, 15])

b = 110
w = -1
alpha = 0.0001

N_ITER = 1000
N_DATA = len(x_lst)

l_sum = 0
l_lst = []
for i in range(1,N_ITER):
    x = x_lst[i%N_DATA]
    y = y_lst[i%N_DATA]

    grad = 2*x*(x*w+b-y)
    w = w - alpha*grad

    yhat = w*x+b
    loss = (y-yhat)**2
    l_sum += loss.item()

    if i % N_DATA == 0:
        l_sum /= N_DATA
        print(f'w={w:0.2f}',
              f'loss={l_sum:0.4f}')
        l_lst.append(l_sum)
        l_sum = 0

plt.plot(range(N_ITER//N_DATA-1), l_lst)
plt.xlabel('iteration')
plt.ylabel('loss')
plt.grid(True)
plt.show()
```
