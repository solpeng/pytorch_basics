```python
import torch
import matplotlib.pyplot as plt

def forward(x):
    y_hat = w*x + b
    return y_hat

def update(x, y):
    global w, b
    grad_w = 2*x*(x*w+b-y)
    grad_b = 2*(b-y+w*x)
    w = w - alpha*grad_w
    b = b - alpha*grad_b

x_lst = torch.Tensor([1, 3, 5, 7, 9])
y_lst = torch.Tensor([100, 75, 60, 25, 15])

w = -1
b = 1
alpha = 0.0005

N_EPOCH = 5000

for epoch in range(N_EPOCH):
    loss = 0
    for x, y in zip(x_lst, y_lst):
        y_hat = forward(x)
        loss += (y-y_hat)**2
        update(x, y)

    if epoch % 10 == 0:
        print(f'epoch={epoch:03}',
              f'w={w:0.2f}, b={b:0.2f}',
              f'loss={loss/len(x_lst):0.5f}')
        
y_hat = [w*x+b for x in x_lst]
plt.plot(x_lst, y_lst, 'o')
plt.plot(x_lst, y_hat, '-*')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(('true', 'predicted'))
plt.grid(True)
plt.show()
```
