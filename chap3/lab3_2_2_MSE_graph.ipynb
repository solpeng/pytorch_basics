```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

def h(z):
    return z

x = torch.Tensor([1, 3, 5, 7, 9])
y_true = torch.Tensor([100, 75, 60, 25, 15])

w_lst = np.linspace(-9, -13, 50)
b_lst = np.linspace(100, 120, 50)

MSE_lst = torch.FloatTensor(len(w_lst),len(b_lst))
for wc in range(len(w_lst)):
    w = w_lst[wc]
    for bc in range(len(b_lst)):
        b = b_lst[bc]
        z = w*x + b
        yhat = h(z)
        mse = sum([(y_true[i]-yhat[i])**2 
                   for i in range(len(x))]) / len(x)
        MSE_lst[wc][bc] = mse

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X, Y = np.meshgrid(w_lst, b_lst)
ax.plot_surface(X, Y, MSE_lst.numpy(),
                       rstride=2,
                       cstride=2,
                       cmap=cm.Blues,
                       linewidth=1,
                       antialiased=True)
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('MSE')
ax.view_init(elev=30,azim=-20)
ax.dist=10
plt.tight_layout()
plt.show()
```
